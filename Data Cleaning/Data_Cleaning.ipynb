```python
import pandas as pd

# Load the data
file_path = 'data/test_data.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
df1 = data.iloc[::2]  # 1st, 3rd, 5th, ...
df2 = data.iloc[1::2] # 2nd, 4th, 6th, ...
length_df1 = len(df1)
length_df2 = len(df2)
total_length = len(data)

if length_df1 == length_df2:
    targets = pd.DataFrame({
        'Saved_df1': df1['Saved'].reset_index(drop=True),
        'Saved_df2': df2['Saved'].reset_index(drop=True)
    })
    preview_df1 = df1.head()
    preview_df2 = df2.head()
    preview_targets = targets.head()
else:   
    #adjusting the length of 'Saved' columns to be equal for the new data
    min_length = min(len(df1), len(df2))
    df1_adjusted = df1.iloc[:min_length]
    df2_adjusted = df2.iloc[:min_length]
    targets = pd.DataFrame({
    'Saved_df1_new': df1_adjusted['Saved'].apply(lambda x: 1 if x == 1 else 0).reset_index(drop=True)
})
    preview_df1 = df1_adjusted.head()
    preview_df2 = df2_adjusted.head()
    preview_targets = targets.head()


preview_df1, preview_df2, preview_targets, (len(df1), len(df2), len(targets))
```

```python
df1 = df1.drop(columns=['Saved','ResponseID'])
df2 = df2.drop(columns=['Saved','ResponseID'])

```

```python
df1.to_csv('df1_test.csv', index=False)
df2.to_csv('df2_test.csv', index=False)
targets.to_csv('targets_test.csv', index=False)
```
