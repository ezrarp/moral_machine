```python
import pandas as pd

# Load the data
file_path = 'sample_data.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
df1 = data.iloc[::2]  # 1st, 3rd, 5th, ...
df2 = data.iloc[1::2] # 2nd, 4th, 6th, ...
length_df1 = len(df1)
length_df2 = len(df2)
total_length = len(data)

if length_df1 == length_df2:
    targets = pd.DataFrame({
        'Saved_df1': df1['Saved'].reset_index(drop=True),
        'Saved_df2': df2['Saved'].reset_index(drop=True)
    })
else:
    #adjusting the length of 'Saved' columns to be equal for the new data
    min_length = min(len(df1), len(df2))
    saved_df1 = df1['Saved'].iloc[:min_length].reset_index(drop=True)
    saved_df2 = df2['Saved'].reset_index(drop=True)
    targets = pd.DataFrame({'Saved_df1': saved_df1, 'Saved_df2': saved_df2})

preview_df1 = df1.head()
preview_df2 = df2.head()
preview_targets = targets.head()

preview_df1, preview_df2, preview_targets, (length_df1, length_df2, total_length)
```
